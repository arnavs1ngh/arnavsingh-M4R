```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read in the results
results = pd.read_csv('data/shallow-x2+2x+1/Shallow_x2+2x+1-10.csv')
# Remove 1st column
results = results.iloc[:,1:]
results = results.groupby('N').mean()

# calculate log of the results
log_results = np.log(results)
log_n = np.log(results.index)

# fit a line to the log results for n > 100
coeffs = np.polyfit(log_n[log_n > 2], log_results['Loss'][log_n > 2], 1)
print(coeffs)

# plot the results
plt.plot(log_n, log_results['Loss'], label='Loss')
# Plot fit line for n > 100
plt.plot(log_n[log_n > 2], coeffs[0]*log_n[log_n > 2] + coeffs[1], label=f"Fit: {coeffs[0]:.2f}x + {coeffs[1]:.2f}")
plt.title("f(x) = x^2 + 2x + 1, Mesh Res = 10")
plt.xlabel("(Log) Width of Network")
plt.ylabel("(Log) Loss")
plt.legend()
plt.show()
```
